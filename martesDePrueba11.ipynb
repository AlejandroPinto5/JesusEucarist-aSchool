```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sb
import plotly.express as px
```

```python
df = pd.read_csv('eleven.csv') 
df['Lastname_Name'] = df['Lastname_Name'].astype('string')
# df['Name'] = df['Name'].astype('string') 
df.dtypes
```

```python
df
```

```python
df.describe(include="all"
                  # percentiles=[0,1]
                  )
```

```python
# df[['Lastname','Physics1', 'Physics2', 'Physics3', 'Physics4', 'Physics5']].trama(x='Lastname', tipo='barra')
# df.plot(x="Lastname", y=["Physics1", "Physics2", "Physics3", "Physics4", "Physics5"], kind="kde")
df.plot.bar(x="Lastname_Name", y=["Physics1", "Physics2", "Physics3", "Physics4", "Physics5"],
            width=1,
            # stacked = 'True',
            alpha=0.6, #Transparency
            subplots=True,
            # legend="bottom"
            sharex=True,
            layout=(5,1),
            figsize=(10,7),
            # title="Cicle No. 1 - 11 Grade",
            legend=False,
            xlabel="Student",
            fontsize=12,
            # rot=45,
            color={"Physics1":"blue","Physics2":"red","Physics3":"green","Physics4":"purple","Physics5":"yellow"}
            )
plt.tight_layout(pad=0.5)
# plt.legend(bbox_to_anchor=(1.0, 1.0))
# plt.show()
```

```python
# dfd = df.drop(['Name'], axis=1)
fig = px.scatter(df, 
              y=["Physics1", "Physics2", "Physics3", "Physics4", "Physics5"], 
              x='Lastname_Name',
              # color='Physics3',
              # facet_col = 'Lastname_Name',
              # facet_col_wrap=5,
              height=600, width=800
              )
fig.update_xaxes(matches=None, showticklabels = True)
fig.show()
```

```python
dfm = df.melt('Lastname_Name', var_name='Physics', value_name='vals_Physics')
dfm
```

```python
dfm.describe(include="all"
                  # percentiles=[0,1]
                  )
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=False, gridspec_kw={'wspace': 0.3})
fig.suptitle('Density - Cicle 1 Ten grade')
sb.set()
sb.kdeplot(data=dfm,ax=axes[0],x="vals_Physics",
           fill="blue",common_norm=False, 
           linewidth=0.5, alpha=.5)
# axes[0].set_title(bulbasaur.name)

sb.kdeplot(data=dfm,ax=axes[1], x="vals_Physics", hue="Physics",
           fill=True,common_norm=False, 
           palette="coolwarm", linewidth=0.1, alpha=.3)
# axes[1].set_title(charmander.name)
sb.kdeplot(data=dfm,ax=axes[2],x="vals_Physics",hue="Lastname_Name",
           legend=False)
median = dfm["vals_Physics"].median()
axes[0].axvline(median, color='r', linestyle='--')
axes[1].axvline(median, color='r', linestyle='--')
axes[2].axvline(median, color='r', linestyle='--')
# a =sb.kdeplot(data=dfm,x="vals_Physics", hue="Physics",
          # fill=True, 
# )
# b=sb.kdeplot(data=dfm)
print(median)
```

```python
ordered_test = dfm.Physics.value_counts().index
t = sb.FacetGrid(dfm, row="Physics", row_order=ordered_test,
                  height=1.7, aspect=4,)
t.map(sb.kdeplot, "vals_Physics")
```

```python
plot=sb.relplot(data=dfm,
          #  kind="violin",
           x="Lastname_Name", y="vals_Physics", hue="Physics", 
           size="Physics", sizes=(15, 200),
           height=7,
           palette="Blues"
           )
plot.set_xticklabels(rotation=90
                     )
```

```python
cicle1 = pd.pivot_table(dfm, values='vals_Physics', 
                              columns='Physics', 
                              index='Lastname_Name')
plt.figure(figsize=(10, 15))
sb.heatmap(cicle1, annot=True,fmt=".4g",cmap='RdYlGn', 
           linewidths=0.1) 
```

```python
# sb.pointplot(y="vals_Physics", x='Physics',  hue = 'Lastname', data=dfdm)
g = sb.FacetGrid(dfm, col="Lastname_Name", col_wrap=4, height=3,sharex=False, despine=True,
                   )
g.map(sb.scatterplot, "Physics", "vals_Physics")
g.set_xticklabels(rotation=45)
g.refline(y=dfm["vals_Physics"].median(), color='red', alpha=0.5)
# print(dfm["vals_Physics"].median)
```
