# ***ADD QUESTION WITH BOOLEAN ANSWER***


```python
dfq = pd.read_csv('tenq.csv') #DataFrame Questions
dfq['competence'] = dfq['competence'].astype('string')
dfq['component'] = dfq['component'].astype('string')
dfq['answer_test'] = dfq['answer_test'].astype('string')
dfq['Details'] = dfq['Details'].astype('string')
dfq['difficulty'] = dfq['difficulty'].astype('string')

dfq.dtypes
```

```python
dfq
```

```python
dft=dfmq.merge(dfq)#Dataframe total
dft.head()
```

```python
dft['C/I'] = dft['answer_test'] == dft['answer_student']
# dft['C/I'] = [i for i in dft if dft['answer_student'] == dft['answer_test']] #C/I is boolean if answer is Correct or Incorrect
dft.head()
```

```python
dft.value_counts('C/I')
```

```python
 df['Good'] = dft.groupby('Lastname')['C/I'].transform('sum') #Count True depends of the student
 df['Wrong'] = dft.groupby('Lastname')['C/I'].transform('count') - dft.groupby('Lastname')['C/I'].transform('sum')#Count false depends of the student

# df['True'] = dft['C/I'].value_counts()[True]
df.head()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5), sharey=False, gridspec_kw={'wspace': 0.3})
fig.suptitle('C/I - Cicle 1 Ten grade')
sb.set()
sb.histplot(data=df,ax=axes[0],y="Good",
           fill="blue",common_norm=False, 
           linewidth=0.5, alpha=.7)

sb.histplot(data=df,ax=axes[1], y="Wrong",
           fill=True,common_norm=False, 
           palette="coolwarm", linewidth=0.1, alpha=.3)

plt.xlabel('Number of Students')

# axes[0].axvline(median, color='r', linestyle='--')
# axes[1].axvline(median, color='r', linestyle='--')
```
